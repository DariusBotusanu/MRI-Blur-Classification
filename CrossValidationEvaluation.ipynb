```python
import pandas as pd
training_logs_cross = [f'Callbacks/{name_0}/training.log',
                       f'Callbacks/{name_1}/training.log',
                       f'Callbacks/{name_2}/training.log',
                       f'Callbacks/{name_3}/training.log',
                       f'Callbacks/{name_4}/training.log'
                       ]

#validation accuracy scores of the crossvalidation
val_df = pd.DataFrame()
train_df = pd.DataFrame()
for i, path in enumerate(training_logs_cross):
  val_df[f'validation {i}'] = pd.read_csv(path, index_col='epoch')['val_accuracy']
  train_df[f'training {i}'] = pd.read_csv(path, index_col='epoch')['accuracy']
```

```python
val_mean_on_crossval = val_df.transpose().describe().loc['mean'] #average validation accuracy on crossvalidation score
val_std_on_crossval = val_df.transpose().describe().loc['std'] #standard deviation of crossvalidation validation accuracy on 

train_mean_on_crossval = train_df.transpose().describe().loc['mean'] #average training accuracy on crossvalidation score
train_std_on_crossval = train_df.transpose().describe().loc['std'] #standard deviation of crossvalidation training accuracy on 
```

```python
df = pd.DataFrame()
df['training_accuracy'] = train_mean_on_crossval
df['validation_accuracy'] = val_mean_on_crossval
```

```python
fig, ax = plt.subplots()
ax.set_title('Average training accuracy and average validation accuracy')
ax.set_xticks(df.index)
ax.set_xlabel('epoch')
ax.set_ylabel('accuracy')
plt1, = ax.plot(df['training_accuracy'], label = 'Training accuracy', color='blue')
plt2, = ax.plot(df['validation_accuracy'], label = 'Validation accuracy', color='orange')
plt3 = ax.fill_between(train_mean_on_crossval.index, train_mean_on_crossval-train_std_on_crossval, train_mean_on_crossval+train_std_on_crossval, alpha=0.3, color='blue')
plt4 = ax.fill_between(val_mean_on_crossval.index, val_mean_on_crossval-val_std_on_crossval, val_mean_on_crossval+val_std_on_crossval, alpha=0.3, color='orange')
plt.legend();
```
